```python
import numpy as np

import pandas as pd







import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv(r"C:\Users\Sonali\Desktop\AirPassengers.csv")

#parse string to datetime type

dataset['Month']=pd.to_datetime(dataset['Month'],infer_datetime_format=True)

Dataset=dataset.set_index(['Month'])
```

```python
Dataset_logScale=np.log(Dataset)

plt.plot(Dataset_logScale)

plt.show()
```

```python
#Plot graph 

plt.xlabel("Date")

plt.ylabel("Number of Air Passengers")

plt.plot(Dataset)

plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition=seasonal_decompose(Dataset_logScale)

trend=decomposition.trend

seasonal=decomposition.seasonal
plt.plot(trend,label='Trend')
```

```python
plt.plot(seasonal,label='Seasonality')
```

```python
Dataset['#Passengers_diff'] = Dataset['#Passengers']- Dataset['#Passengers'].shift(7)
Dataset['#Passengers_diff'].dropna().plot()
```
